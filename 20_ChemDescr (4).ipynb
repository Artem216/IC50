```python
data=df.drop(['Title', 'SMILES'], axis=1)
```

```python
data.head()
```

```python
cols=data.columns[:42]
data=data[cols]
data.head()
```

```python
data=data.dropna()
nan_counts = data.isnull().sum()
nan_counts.sum()
```

```python
y=data['IC50']
data=data.drop(['IC50', 'Unnamed: 0'], axis=1)
```

```python
!pip install torch
import torch
```

```python
if torch.cuda.is_available():
    device = torch.device('cpu')
    print('CUDA доступен. Используем устройство:', device)
else:
    device = torch.device('cpu')
    print('CUDA недоступен. Используем устройство:', device)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    data.values,
    y.values,
    test_size=0.1,
    shuffle=True)

X_train = torch.FloatTensor(X_train).to(device)
X_test = torch.FloatTensor(X_test).to(device)
y_train = torch.FloatTensor(y_train).to(device)
y_test = torch.FloatTensor(y_test).to(device)
```

```python
X_train.shape
```

```python
import torch


class WineNet(torch.nn.Module):
    def __init__(self, n_hidden_neurons):
        super(WineNet, self).__init__()

        self.fc1 = torch.nn.Linear(40, n_hidden_neurons)
        self.activ1 = torch.nn.Sigmoid()
        self.fc2 = torch.nn.Linear(n_hidden_neurons, 1)

    def forward(self, x):
        x = self.fc1(x)
        x = self.activ1(x)
        x = self.fc2(x)
        return x


wine_net = WineNet(50).to(device)
```

```python
loss = torch.nn.MSELoss()

optimizer = torch.optim.Adam(wine_net.parameters(),
                             lr=1.0e-3)
```

```python
import numpy as np
batch_size = 256

for epoch in range(5000):
    order = np.random.permutation(len(X_train))
    for start_index in range(0, len(X_train), batch_size):
        optimizer.zero_grad()

        batch_indexes = order[start_index:start_index+batch_size]

        x_batch = X_train[batch_indexes]
        y_batch = y_train[batch_indexes]

        preds = wine_net.forward(x_batch)
        loss_value = loss(preds.squeeze(1), y_batch)

        loss_value.backward()
        optimizer.step()

    if epoch % 100 == 0:
        test_preds = wine_net.forward(X_test)
        print('metric: ', (((test_preds.squeeze(-1) - y_test)**2).sum()/len(y_test)).data)
        print('loss_value: ', loss_value.data)
```

```python
test_preds = wine_net.forward(X_test)
for i in range (0,len (test_preds)):
  print(f'predicted: {round(test_preds[i].item(), 2)}, real: {round(y_test[i].item(), 2)}, difference: {round((abs(test_preds[i]-y_test[i])).item(), 2)}')
```
